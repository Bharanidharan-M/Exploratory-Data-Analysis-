```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
data=pd.read_csv("/content/diabetes.csv")
data.head()
```

```python
data.isnull().sum()
```

MATPLOTLIB

```python
x=data["BMI"]
y=data["Age"]
```

```python
plt.title("scatter plot")
plt.scatter(x,y)
```

```python
z=data["Age"]
plt.title("Line plot")
plt.plot(z)
```

```python
x=data["DiabetesPedigreeFunction"]
plt.title("Histogram")
plt.hist(x)
```

```python
x=data["Age"]
y=data["Insulin"]
plt.title("Bar")
plt.bar(x,y)
```
